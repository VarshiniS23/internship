```python
x=df.iloc[:,0:1].values
y=df.iloc[:,1].values
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.predict([[7.6]])
```

```python
y_p=lr.predict(x)
y_p
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,color="green",label="Achual_Values")
plt.plot(x,y_p,color="purple",label="Predicted_Values")
plt.legend()
plt.xlabel("Amount",fontsize=14)
plt.ylabel("Class",fontsize=14)
plt.title("Linear Regression to predict credit card fraud detection")
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y,y_p)      #R2 score : Co-efficient of determination
```

```python
mean_squared_error(y,y_p)
```
